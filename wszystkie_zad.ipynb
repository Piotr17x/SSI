```python
import pandas as pd
import numpy as np


class FileReader:
    def __init__(self, path: str, header: list, sep: str = '\t'):
        self.path = path
        try:
            self.df = pd.read_csv(
                self.path,
                header=None,
                sep=sep,
                names=header
            )
        except FileNotFoundError:
            print('Niepoprawna nazwa pliku')

    def get_df(self) -> pd.core.frame.DataFrame:
        return self.df

    def get_row(self, row_numb: int) -> pd.core.series.Series:
        return self.df.iloc[row_numb]
```

k-means

```python
from matplotlib.colors import CSS4_COLORS
import random


colors = list(CSS4_COLORS.values())
random.shuffle(colors)
```

```python
def euc_distance2d(p1, p2):
    return np.sqrt(pow(p2.x-p1.x, 2)+pow(p2.y-p1.y, 2))


def plot_k_means(V, df, legend=''):
    for idm, m in V.iterrows():
        plt.scatter(
            [df.loc[df['group'] == idm]['x']],
            [df.loc[df['group'] == idm]['y']],
            color=colors[idm],
        )
        plt.scatter(
            m['x'],
            m['y'],
            color=colors[idm],
            marker='D',
            edgecolors='#000000',
            label=f'class {idm}',
        )
    if legend:
        plt.legend(loc='upper right')
    plt.show()


def k_means(m: int, iters: int, df, plot_iters=None):
  if not plot_iters:
    plot_iters = [0, iters-1]
  V = df.sample(m)
  V = V.reset_index()
  df['group'] = -1
  for x in range(iters):
      for ids, s in df.iterrows():
          dist = np.inf
          for idm, m in V.iterrows():
              if dist > euc_distance2d(s, m):
                  dist = euc_distance2d(s, m)
                  df.loc[ids, 'group'] = idm

      if x in plot_iters:
          plt.title(f'k-means rozmieszczenie iter. {x+1}')
          plot_k_means(V, df, True if x == plot_iters[0] else False)

      for idm, m in V.iterrows():
          group = df.loc[df['group'] == idm]
          if len(group.index):
              V.loc[idm, 'x'] = group['x'].mean()
              V.loc[idm, 'y'] = group['y'].mean()
```

strategia 1+1

```python
def funkcja_przystosowania(x):
    return np.sin(x/10)*np.sin(x/200)


def alg_1plus1(rozrzut, wsp_przyrostu, l_iteracji, zakres_zmiennosci=(0, 100)):
    x = random.uniform(zakres_zmiennosci[0], zakres_zmiennosci[1])
    y = funkcja_przystosowania(x)
    for i in range(l_iteracji):
        x_pot = x + random.uniform(-rozrzut, rozrzut)
        if x_pot > zakres_zmiennosci[1]:
            x_pot = zakres_zmiennosci[1]
        elif x_pot < zakres_zmiennosci[0]:
            x_pot = zakres_zmiennosci
        y_pot = funkcja_przystosowania(x_pot)
        if y_pot >= y:
            x = x_pot
            y = y_pot
            rozrzut *= wsp_przyrostu
        else:
            rozrzut /= wsp_przyrostu
        print(f'Iteracja #{i+1}\n'
              f'x = {x}\n'
              f'y = {y}\n'
              f'rozrzut = {rozrzut}\n'
              f'+=+=+=+=+=+')
        plt.title('Algorytm 1+1')
        plt.plot(x, y, marker='o', color='g')
    x = np.arange(0, 100, 0.5)
    plt.plot(x, funkcja_przystosowania(x), label='f(x)=sin(x/10)*sin(x/200)')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='upper left')
    plt.show()
```

FCM

```python
colors_for_3 = ['r', 'g', 'b']

class FCM:
    def __init__(self, df, c_numb, m=2):
        self.m = m
        self.c_numb = c_numb
        self.df = df
        self.V = df.sample(c_numb)
        self.V = self.V.reset_index()

        for group in range(c_numb):
            self.df[f'U_{group}'] = -1
            self.df[f'D_{group}'] = 0

        for ids, s in df.iterrows():
            r = [random.random() for group in range(c_numb)]
            for group in range(c_numb):
                df.loc[ids, f'U_{group}'] = r[group]/sum(r)

    def set_centroids(self):
        for idm in range(len(self.V.index)):
            u = sum([pow(self.df.iloc[ids][f'U_{idm}'], 2) for ids in range(len(self.df.index))])
            self.V.loc[idm, 'x'] = sum([pow(self.df.iloc[ids][f'U_{idm}'], 2)*self.df.iloc[ids]['x'] for ids in range(len(self.df.index))])/u
            self.V.loc[idm, 'y'] = sum([pow(self.df.iloc[ids][f'U_{idm}'], 2)*self.df.iloc[ids]['y'] for ids in range(len(self.df.index))])/u

    def update_membership(self):
        for ids in range(len(self.df.index)):
            for idm, m in self.V.iterrows():
                self.df.loc[ids, f'D_{idm}'] = self.euc_distance2d(self.df.iloc[ids], m) if self.euc_distance2d(self.df.iloc[ids], m) > 0.0005 else 0.0005
            for idm, m in self.V.iterrows():
                self.df.loc[ids, f'U_{idm}'] = pow(pow(sum([pow(self.df.iloc[ids][f'D_{idm}'], 2)/pow(self.df.iloc[ids][f'D_{group}'], 2) for group in range(self.c_numb)]), 1/(self.m-1)), -1)

    @staticmethod
    def euc_distance2d(p1, p2):
        return np.sqrt(pow(p2.x-p1.x, 2)+pow(p2.y-p1.y, 2))

    def iterate(self, x, plot_iters=None):
        if not plot_iters:
            plot_iters = [0, x-1]
        for i in range(x):
            self.set_centroids()
            if i in plot_iters:
                plt.title(f'FCM rozmieszczenie iter. {i+1}')
                plot_fcm(self.V, self.df, True if i == plot_iters[0] else False)
            self.update_membership()


def plot_fcm(V, df, legend=''):
    if len(V.index) == 3:
      for ids, s in df.iterrows():
          plt.scatter(
              s['x'],
              s['y'],
              color=(s['U_0'], s['U_1'], s['U_2'])
          )
      for idm, m in V.iterrows():
          plt.scatter(
              m['x'],
              m['y'],
              color=colors_for_3[idm],
              marker='D',
              edgecolors='#000000',
              label=f'class {idm}',
          )
    else:
      for ids, s in df.iterrows():
          max_id = df.iloc[ids][[f'U_{idm}' for idm in range(len(V.index))]].idxmax()
          max_id = int(max_id[2])
          plt.scatter(
              s['x'],
              s['y'],
              color=colors[max_id],
          )
      for idm, m in V.iterrows():
          plt.scatter(
              m['x'],
              m['y'],
              color=colors[idm],
              marker='D',
              edgecolors='#000000',
              label=f'class {idm+1}',
          )

    if legend:
        plt.legend(loc='upper left')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()
```

my+lambda

```python
import copy


X, Y = np.meshgrid(np.linspace(0, 100, 100), np.linspace(0, 100, 100))
Z = np.sin(X * 0.05) + np.sin(Y * 0.05) + 0.4 * np.sin(X * 0.15) * np.sin(Y * 0.15)


def funkcja_przystosowania(x):
    return np.sin(x[0]*0.05)+np.sin(x[1]*0.05)+0.4*np.sin(x[0]*0.15)*np.sin(x[1]*0.15)


def my_lambda(my, lmbd, turniej_rozmiar, mutacja_poziom, iteracje_liczba, plot_iters=None):
    if not plot_iters:
        plot_iters = [0, iteracje_liczba-1]
    rodz = []
    for x in range(my):
        rodz.append([random.uniform(0, 100), random.uniform(0, 100)])
    for x in range(iteracje_liczba):
        pot = []
        while len(pot) < lmbd:
            oss_turniej = [random.sample(copy.deepcopy(rodz), turniej_rozmiar)][0]
            os_n = max(oss_turniej, key=lambda z: funkcja_przystosowania(z))
            os_n[0] += random.uniform(-mutacja_poziom, mutacja_poziom)
            os_n[1] += random.uniform(-mutacja_poziom, mutacja_poziom)
            check_range(os_n)
            pot.append(os_n)
        if x in plot_iters:
            func = f'f(X,Y) = sin(X * 0.05) + sin(Y * 0.05) + 0.4 * sin(X * 0.15) * sin(Y * 0.15)\n' if x == plot_iters[0] else ''
            plt.title(f'µ + lambda iter. {x+1}\n'
                      f'{func}'
                      f'najlepszy wynik to: {round(max([funkcja_przystosowania(z) for z in rodz + pot]), 4)}')
            plot = plt.contour(X, Y, Z, levels=13)
            plt.clabel(plot, inline=True, fontsize=10)
            plt.xlabel('x')
            plt.ylabel('y')
            plt.scatter([x[0] for x in rodz], [x[1] for x in rodz], marker='*', label='pula rodz.')
            plt.scatter([x[0] for x in pot], [x[1] for x in pot], marker='o', label='pula pot.')
            if x == plot_iters[0]:
                plt.legend(loc='upper right')
            plt.show()
        rodz = sorted(rodz+pot, key=lambda z: funkcja_przystosowania(z), reverse=True)
        rodz = rodz[:my]


def check_range(os):
    if os[0] < 0:
        os[0] = 0
    elif os[0] > 100:
        os[0] = 100
    if os[1] < 0:
        os[1] = 0
    elif os[1] > 100:
        os[1] = 100
```

```python
from PIL import Image


def euc_distance2d(p1, p2):
    return np.sqrt(pow(p2[0]-p1[0], 2)+pow(p2[1]-p1[1], 2))


def dopasuj(test_znak, baza_wzor):
    if not len(baza_wzor):
        print('brak znaków w bazie wzorcowej')
        return
    miara = -np.inf
    wynik = 0
    for idb, znak in enumerate(baza_wzor):
        miara_p_obustronna = -(miara_niepodobieństwa(test_znak, znak) + miara_niepodobieństwa(znak, test_znak))
        if miara_p_obustronna > miara:
            miara = miara_p_obustronna
            wynik = idb
    return test_znak, baza_wzor[wynik]


def miara_niepodobieństwa(b1, b2):
    miara = 0
    for idr, row in enumerate(b1):
        for idc, point in enumerate(row):
            if point:
                odl_min = np.inf
                for idr_2, row2 in enumerate(b2):
                    for idc_2, point2 in enumerate(row2):
                        if point2:
                            odl_min = min(
                                odl_min,
                                euc_distance2d(
                                    (idr, idc),
                                    (idr_2, idc_2),
                                )
                            )
                miara += odl_min
    return miara


def read_matrix(filename, x, y):
    f = open(filename, 'r')
    baza = []
    for line in f.read().splitlines():
        mat = np.array([int(i) for i in line])
        mat.resize(x, y)
        baza.append(mat)
    return baza


def get_color(i,j,mat, cell_size):
    value = mat[int(np.floor(j/cell_size))][int(np.floor(i/cell_size))]
    color = 0 if value == 1 else 255
    return (color, color, color)


def display_matrix(mat, cell_size):
    img = Image.new( 'RGB', (len(mat[0])*cell_size+1, len(mat)*cell_size+1), "black") # Create a new black image
    pixels = img.load()
    for i in range(img.size[0]):
        for j in range(img.size[1]):
            pixels[i,j] = get_color(i, j, mat, cell_size) if i % cell_size !=0 and j % cell_size != 0 else (255, 0, 0)
    display(img)

```

Hopfield

```python
class Hopfield:
    def __init__(self, h, w):
        self.h = h
        self.w = w
        self.n = h * w
        self.weights = np.zeros((self.n, self.n))

    def naucz_obraz(self, wzor):
        flat = wzor.flatten()
        for i in range(self.n):
            for j in range(self.n):
                if j == i:
                    continue
                self.weights[i][j] += (flat[i]*flat[j])/self.n

    def rozpoznaj_obraz(self, obraz):
        flat = obraz.flatten()
        for i in range(self.n):
            suma = 0
            for j in range(self.n):
                if i == j:
                    continue
                suma += flat[j]*self.weights[i][j]
            flat[i] = 1 if suma >= 0 else -1
        rozp_obraz = np.resize(flat, (self.h, self.w))
        if np.array_equal(rozp_obraz, obraz):
            return False
        return rozp_obraz

    @staticmethod
    def read_matrix(filename, x, y):
        f = open(filename, 'r')
        baza = []
        for line in f.read().splitlines():
            mat = np.array([int(i) for i in line.replace('-1', '0')])
            mat[mat == 0] = -1
            mat.resize(x, y)
            baza.append(mat)
        return baza
```

```python
row, col = 5, 5
baza_test = Hopfield.read_matrix('hopfield_test.txt', row, col)
baza_wzor = Hopfield.read_matrix('hopfield_wzor.txt', row, col)
hopfield = Hopfield(row, col)
for wzor in baza_wzor:
    hopfield.naucz_obraz(wzor)
print(f'Znak testowy:')
id_do_zmiany = 0
display_matrix(baza_test[id_do_zmiany], 45)
zmiana = hopfield.rozpoznaj_obraz(baza_test[id_do_zmiany])
if type(zmiana) != bool:
    print('został zmieniony na:')
    display_matrix(zmiana, 45)
else:
    print('nie został zmieniony')
```

Fireflies

```python
import random


_X, _Y = np.meshgrid(np.linspace(0, 100, 100), np.linspace(0, 100, 100))
_Z = np.sin(_X * 0.05) + np.sin(_Y * 0.05) + 0.4 * np.sin(_X * 0.15) * np.sin(_Y * 0.15)


def euc_distance2d(p1, p2):
    return np.sqrt(pow(p2[0]-p1[0], 2)+pow(p2[1]-p1[1], 2))


def funkcja_przystosowania(x):
    return np.sin(x[0]*0.05)+np.sin(x[1]*0.05)+0.4*np.sin(x[0]*0.15)*np.sin(x[1]*0.15)


def firefly(N, beta_0, gamma, mu, iteracje_liczba, plot_iters=None):
    if not plot_iters:
        plot_iters = [0, iteracje_liczba-1]
    X = np.array([[random.uniform(0, 100), random.uniform(0, 100)] for x in range(N)])
    for x in range(iteracje_liczba):
        if x in plot_iters:
            func = f'f(X,Y) = sin(X * 0.05) + sin(Y * 0.05) + 0.4 * sin(X * 0.15) * sin(Y * 0.15)\n' if x == plot_iters[0] else ''
            plt.title(f'Fireflies iter. {x+1}\n'
                      f'{func}'
                      f'najlepszy wynik to: {round(max([funkcja_przystosowania(z) for z in X]), 4)}')
            plot = plt.contour(_X, _Y, _Z, levels=13)
            plt.clabel(plot, inline=True, fontsize=10)
            plt.xlabel('x')
            plt.ylabel('y')
            plt.scatter([z[0] for z in X], [z[1] for z in X], marker='*', label='fireflies')
            if x == plot_iters[0]:
                plt.legend(loc='upper right')
            plt.show()
        for ida, a in enumerate(X):
            for idb, b in enumerate(X):
                if funkcja_przystosowania(b) > funkcja_przystosowania(a):
                    beta = beta_0 * np.exp(-gamma*pow(euc_distance2d(a, b), 2))
                    X[ida] = X[ida] + beta * (X[idb] - X[ida])
            X[ida] = X[ida] + np.array([random.uniform(-(100 * mu), 100 * mu), random.uniform(-(100 * mu), 100 * mu)])
            check_range(X[ida])


def check_range(os):
    if os[0] < 0:
        os[0] = 0
    elif os[0] > 100:
        os[0] = 100
    if os[1] < 0:
        os[1] = 0
    elif os[1] > 100:
        os[1] = 100
```

```python
x_min, x_max, x_zasieg = -100, 100, 200
y_min, y_max, y_zasieg = -100, 0, 100
rampa_lewy, rampa_prawy, rampa_dol, rampa_gora = -30, 30, -10, 0
rampa_kat_docelowy_min = -20
rampa_kat_docelowy_max = 20
obrot_min, obrot_max = -20, 20
kat_min, kat_max = -180, 180
kat_zasieg = 360
ruch_skok = 5


def skret_prawo_dol(kat):
    return np.exp(-((kat-135)/50)**2)


def skret_prawo_gora(kat):
    return np.exp(-((kat-45)/50)**2)


def lewa_czesc(x):
    return np.exp(-((x+np.abs(x_min+rampa_lewy)/2)/(np.abs(x_min-rampa_lewy)/2))**2)


def prawa_czesc(x):
    return np.exp(-((x-np.abs(x_max+rampa_prawy)/2)/(np.abs(x_max-rampa_prawy)/2))**2)


def srodek(x):
    return np.exp(-((x-np.abs(rampa_lewy+rampa_prawy)/2)/(np.abs(rampa_lewy-rampa_prawy)/2))**2)


def skret_lewo_dol(kat):
    return np.exp(-((kat + 135) / 50) ** 2)


def skret_lewo_gora(kat):
    return np.exp(-((kat + 45) / 50) ** 2)


def gora(y):
    return np.exp(-((y + np.abs(y_max + y_min)/4) / 30) ** 2)


def dol(y):
    return np.exp(-((y + np.abs(y_max + y_min)/4*3) / 30) ** 2)


def get_skret(x1, y, kat):
    skret = (
            lewa_czesc(x1) * skret_prawo_dol(kat) * gora(y) * 0
            + lewa_czesc(x1) * skret_prawo_gora(kat) * gora(y) * 20
            + lewa_czesc(x1) * skret_prawo_dol(kat) * dol(y) * -20
            + lewa_czesc(x1) * skret_prawo_gora(kat) * dol(y) * 20
            + srodek(x1) * skret_prawo_gora(kat) * -20
            + srodek(x1) * skret_lewo_gora(kat) * 20
            + prawa_czesc(x1) * skret_prawo_gora(kat) * gora(y) * -20
            + prawa_czesc(x1) * skret_prawo_dol(kat) * gora(y) * 0
            + prawa_czesc(x1) * skret_prawo_dol(kat) * dol(y) * 20
            + prawa_czesc(x1) * skret_prawo_gora(kat) * dol(y) * -20
    ) / (
            lewa_czesc(x1) * skret_prawo_dol(kat) * gora(y)
            + lewa_czesc(x1) * skret_prawo_gora(kat) * gora(y)
            + lewa_czesc(x1) * skret_prawo_dol(kat) * dol(y)
            + lewa_czesc(x1) * skret_prawo_gora(kat) * dol(y)
            + srodek(x1) * skret_prawo_gora(kat)
            + srodek(x1) * skret_lewo_gora(kat)
            + prawa_czesc(x1) * skret_prawo_gora(kat) * gora(y)
            + prawa_czesc(x1) * skret_prawo_dol(kat) * gora(y)
            + prawa_czesc(x1) * skret_prawo_dol(kat) * dol(y)
            + prawa_czesc(x1) * skret_prawo_gora(kat) * dol(y)
    )
    return skret


class PojazdPolozenie:
    def __init__(self, x, y, kat):
        self.x = x
        self.y = y
        self.kat = kat
        self.start_kat = kat
        self.road = [[x], [y]]
        plt.scatter(x, y, marker='D', label='start', color='r')

    def zwroc_obrot(self):
        obrot = get_skret(self.x, self.y, self.kat)
        self.kat += obrot

    def ruch(self):
        self.x += (ruch_skok * np.sin(np.radians(self.kat / np.pi)*np.pi))
        self.y += (ruch_skok * np.cos(np.radians(self.kat / np.pi)*np.pi))
        self.road[0].append(self.x)
        self.road[1].append(self.y)

    # -1 = lose; 1 = win; 0 = play
    def check_status(self):
        if x_min <= self.x <= x_max:
            if y_min <= self.y <= y_max:
                if rampa_lewy <= self.x <= rampa_prawy and rampa_dol <= self.y <= rampa_gora and rampa_kat_docelowy_min <= self.kat <= rampa_kat_docelowy_max:
                    return 1
                return 0
        return -1

    def show_map(self):
        wynik = 'zwycięstwo' if pojazd.check_status() == 1 else 'porażka'
        plt.plot(self.road[0], self.road[1], label='droga', color='b')
        plt.scatter(self.road[0][-1], self.road[1][-1], label='koniec', color='g')
        plt.plot(
            [rampa_lewy, rampa_prawy, rampa_prawy, rampa_lewy, rampa_lewy],
            [rampa_gora, rampa_gora, rampa_dol, rampa_dol, rampa_gora],
            linestyle='dotted',
            color='r',
            label='cel'
        )
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend(loc='upper right')
        plt.title(f'Sterownik rozmyty\n'
                  f'start:[{self.road[0][0]}, {self.road[1][0]}] kąt: {self.start_kat}°\n'
                  f'wynik: {wynik}')
        ax = plt.gca()
        ax.set_ylim([-100, 0])
        ax.set_xlim([-100, 100])
        plt.grid()
        plt.show()
```

```python
pojazd = PojazdPolozenie(-90, -30, 0)
while pojazd.check_status() == 0:
    pojazd.zwroc_obrot()
    pojazd.ruch()
pojazd.show_map()

# x=np.arange(-100,100,1)
# plt.plot(x,np.exp(-((x-np.abs(0)/2)/50)**2))
# plt.ylabel('Y')
# plt.show()
```
